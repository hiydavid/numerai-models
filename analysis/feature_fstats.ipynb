# Compute era-wise feature f-stats to target

```python
# load lib
from sklearn.feature_selection import f_regression
```

```python
# function for calculating fstats
def calculate_fstats(df):
    fstats, p_values = f_regression(df[features], df[TARGET_COL])
    return fstats
```

```python
# get feature fstats per era
training_era_fstats = training.groupby(["era"]).apply(lambda x: calculate_fstats(x))
validation_era_fstats = validation.groupby(["era"]).apply(lambda x: calculate_fstats(x))
```

```python
# make into fstats matrix (training)
training_era_fstats_list = []

for i in training_era_fstats.index:
    tdf = pd.DataFrame(training_era_fstats[i]).T
    training_era_fstats_list.append(tdf)

training_era_fstats_df = pd.concat(training_era_fstats_list)
training_era_fstats_df.columns = features
training_era_fstats_df.index = training_era_fstats.index

training_era_fstats_df.head()
```

```python
# make into fstats matrix (validation)
validation_era_fstats_list = []

for i in validation_era_fstats.index:
    vdf = pd.DataFrame(validation_era_fstats[i]).T
    validation_era_fstats_list.append(vdf)

validation_era_fstats_df = pd.concat(validation_era_fstats_list)
validation_era_fstats_df.columns = features
validation_era_fstats_df.index = validation_era_fstats.index

validation_era_fstats_df.head()
```

---
# Get top features by average f-stats

```python
# get top training features
top_training_features = pd.DataFrame(training_era_fstats_df.mean().sort_values(ascending=False))
top_training_features = top_training_features.rename(columns = {0: "training"})
top_training_features
```

```python
# get top training features
top_validation_features = pd.DataFrame(validation_era_fstats_df.mean().sort_values(ascending=False))
top_validation_features = top_validation_features.rename(columns = {0: "validation"})
top_validation_features
```

```python
# get scatter
avg_era_fstats = top_training_features.merge(top_validation_features, left_index=True, right_index=True)
avg_era_fstats.plot.scatter(x="training", y="validation", figsize=(7, 7));
```

```python
# select features
avg_era_fstats["average"] = (avg_era_fstats["training"] + avg_era_fstats["validation"]) / 2
avg_era_fstats["average"].sort_values(ascending=False).head(100)
```

```python
# save as features
top_fstats_features = {
    "top_50_features" : avg_era_fstats["average"].sort_values(ascending=False).head(50).index.tolist(),
    "top_100_features" : avg_era_fstats["average"].sort_values(ascending=False).head(100).index.tolist(),
    "top_200_features" : avg_era_fstats["average"].sort_values(ascending=False).head(200).index.tolist(),
    "top_500_features" : avg_era_fstats["average"].sort_values(ascending=False).head(500).index.tolist()
}
```

```python
# save files
with open("data/top_fstats_features.json", "w") as outfile:
    json.dump(top_fstats_features, outfile, indent = 4)
```

# Test predictiveness of top features

```python
# open top features
with open("data/top_fstats_features.json", "r") as f:
    top_fstats_features = json.load(f)
```

```python
# get features
top50 = top_fstats_features["top_50_features"]
top100 = top_fstats_features["top_100_features"]
top200 = top_fstats_features["top_200_features"]
top500 = top_fstats_features["top_500_features"]

len(top50), len(top100), len(top200), len(top500)
```

```python
# function to train and evaluate models
def run_lgbm(features, target=TARGET_COL):
    params = {"n_estimators": 2000,
              "learning_rate": 0.01,
              "max_depth": 5,
              "num_leaves": 2 ** 5,
              "colsample_bytree": 0.1}
    model = LGBMRegressor(**params)
    model.fit(
        training[features],
        training[target]
    )
    return model.predict(validation[features])
```

```python
# get preds from each feature set
preds_list = []
for i in [top50, top100, top200, top500]:
    preds_list.append(run_lgbm(features=i))
```

```python
# combine preds
preds_all = pd.DataFrame(preds_list).T
preds_all.columns = ["top50", "top100", "top200", "top500"]
preds_all.index = validation.index
preds_all["era"] = validation[ERA_COL]
preds_all["target"] = validation[TARGET_COL]

preds_all.head()
```

---
# Calculate per-era correlations

```python
# era corr function
corr_list = []
for i in ["top50", "top100", "top200", "top500"]:
    r = preds_all.groupby("era")[["target", i]].corr(method="spearman").iloc[0::2, -1]
    corr_list.append(r)
corr_all = pd.concat(corr_list, axis=1).dropna().reset_index()
corr_all = corr_all.drop("level_1", axis=1).set_index("era")

corr_all.head()
```

```python
# plot corr
corr_all.rolling(52 * 2).mean().plot(title="Per-Era Corr", figsize=(10,5));
```

```python
# get sharpe ratios
corr_all.mean() / corr_all.std()
```
